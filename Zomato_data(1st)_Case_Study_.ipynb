Details about Zomato Dataset

 * **Restaurant Id :** Unique id of every restaurant across various cities of the world
 * **Restaurant Name :** Name of the restaurant
 * **Country Code :** Country in which restaurant is located
 * **City:** City in which restaurant is located
 * Address: Address of the restaurant
 * Locality: Location in the city
 * Locality Verbose: Detailed description of the locality
 * Longitude: Longitude coordinate of the restaurant's location
 * Latitude: Latitude coordinate of the restaurant's location
 * Cuisines: Cuisines offered by the restaurant
 * Average Cost for two: Cost for two people in different currencies (local currency)
 * Currency: Currency of the country
 * Has Table booking: yes/no
 * Has Online delivery: yes/ no
 * Is delivering: yes/ no
 * Switch to order menu: yes/no
 * Price range: range of price of food
 * Aggregate Rating: Average rating out of 5
 * Rating color: depending upon the average rating color
 * Rating text: text on the basis of rating of rating
 * Votes: Number of ratings given
 * Switch to order menu is NO for all observations, hence it's of no value.
 * Price range ranges from one to four, with four being premium-priced restaurants.

5. Compare the price range for restaurants that have table booking vs that have online
booking. Plot suitable graphs.

```python
# For Comparison the price range for restaurants that have table booking vs that have online booking I exclude these rows which contain 'Yes' for both. Now we create
# a clear comparion on it.
compare_df = zm[~((zm['Has Table booking'] == 'Yes') & (zm['Has Online delivery'] == 'Yes'))]
```

```python
# fetch the data to to analyze the price range distribution for restaurants that have table booking available

pricerangefor_Table_Booking = compare_df[compare_df['Has Table booking'] == 'Yes']['Price range'].value_counts().sort_index().reset_index(name = 'Count')
pricerangefor_Table_Booking = pricerangefor_Table_Booking.rename(columns = {'index' : 'Price Range'})
pricerangefor_Table_Booking
```

```python
# fetch the data to to analyze the price range distribution for restaurants that Has Online delivery available

pricerangefor_Online_delivery = compare_df[compare_df['Has Online delivery'] == 'Yes']['Price range'].value_counts().sort_index().reset_index(name = 'Count')
pricerangefor_Online_delivery = pricerangefor_Online_delivery.rename(columns = {'index' : 'Price Range'})
pricerangefor_Online_delivery
```

```python
# merge two DataFrames (pricerangefor_Table_Booking and pricerangefor_Online_delivery) based on the 'Price Range' column.

pricerange_df = pd.merge(pricerangefor_Table_Booking,pricerangefor_Online_delivery, on='Price Range',suffixes=('_Dilhi NCR' , '_Rest of India') , how = 'outer').fillna(0)
pricerange_df
```

```python
# Here I use The melt() function to reshape the DataFrame from wide format to long format, which is more suitable for Seaborn plotting.
# It transforms the DataFrame by combining multiple columns into key-value pairs.

df_melted = pd.melt(pricerange_df, id_vars='Price Range', var_name='Region', value_name='Count')
df_melted
```

```python
# plot a bar plot with the help of seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Price Range', y='Count', hue='Region', data=df_melted, palette='husl')
plt.title('Restaurant Count by Price Range in Delhi NCR and Rest of India')
plt.xlabel('Price Range')
plt.ylabel('Count of Restaurants')
plt.show()
```

**Analysis :**

 * The resulting bar plot visually compares the count of restaurants in Delhi NCR and the Rest of India for each price range.
 * Price Range 2 has a substantial presence in both regions, with a higher count in the Rest of India.
 * Price Range 3 is more dominant in Delhi NCR compared to the Rest of India.
 * Price Range 4 has a significant presence in Delhi NCR but is limited in the Rest of India.
 * Price Range 1 has no representation in Delhi NCR but is prevalent in the Rest of India.

This analysis provides insights into how the distribution of restaurants in different price ranges varies based on whether they offer table booking or online delivery. Adjustments to business strategies can be made based on these regional preferences
