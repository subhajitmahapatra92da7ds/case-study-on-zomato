 ## 1.1.  Number of Votes given Restaurant

```python
# group and count data based on the 'Votes' and 'Aggregate rating' columns.
groupeddata_df = zm.groupby('Votes')['Aggregate rating'].value_counts().reset_index(name='count')
groupeddata_df
```

```python
# since for 0 to 3 votes Aggregate rating is 'zero' so it cannot give any insigths. Therefore I remove this rows which contains 'zero' ratings.
groupeddata_df = groupeddata_df.iloc[4:]
groupeddata_df
```

```python
# Create a scatter plot to visualize the relationship between 'Votes' and 'Aggregate rating'.

plt.figure(figsize=(10,7))

sns.scatterplot(data = groupeddata_df , x = 'Votes' , y = 'Aggregate rating', size='count' , alpha=0.9 , color = 'Blue')
plt.title('Effect of Number of Votes on Restaurant Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')

plt.show()
```

**Explanation of using perticular plot for this question :**

Here I use Scatter plotto analyze the relationship between the number of votes and aggregate ratings, considering the frequency of different combinations.
A scatter plot is well-suited for visualizing relationships between two continuous variables (votes and ratings). Size encoding adds an extra dimension, representing the count of occurrences for each combination of votes and ratings. This allows for a quick identification of patterns and trends.

Scatter plots are effective for visualizing the distribution of data points and identifying trends.

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Votes', y='Aggregate rating', data=groupeddata_df)
plt.title('Average Rating for Different Numbers of Votes')
plt.xlabel('Votes')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Create a heatmap to visualize the correlation matrix between 'Votes', 'Aggregate rating', and 'count'.
correlation_matrix = groupeddata_df[['Votes', 'Aggregate rating', 'count']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

 ## 1.2.  Restaurant serving more number of cuisines.

```python
# Create a new column in the exiting dataframe by calcutate the number of cuisines serve by each resturants
zm['Number of Cuisine'] = zm['Cuisines'].apply(lambda x : len(str(x).split(',')))
zm.head(2)
```

```python
# group a DataFrame named zm by the column 'Number of Cuisine' and then calculating the mean of the 'Aggregate rating' for each group.
new_df = zm.groupby('Number of Cuisine')['Aggregate rating'].mean().reset_index(name='Avg rating')
new_df
```

```python
# create a bar plot to visualize the effect of the number of cuisines on average user rating.
plt.figure(figsize=(10,6))
sns.barplot(data = new_df , x = 'Number of Cuisine' , y = 'Avg rating' , color = 'Blue')
plt.title('Effect of Number of Cuisines on User Rating')
plt.show()
```

**Explanation of using perticular plot for this question :**

Here I use Bar Plot to show the average user rating for different numbers of cuisines served. Bar plots are suitable for displaying the mean or average of a numerical variable (in this case, user rating) for different categories (number of cuisines). They allow for a clear comparison between different categories, making it easy to identify trends or patterns.The bar plot clearly illustrates how the average user rating changes as the number of cuisines served increases. This visual representation helps viewers quickly grasp the relationship between these two variables.

```python
# Group by the number of cuisines and calculate mean rating and total votes
rating_by_cuisines = zm.groupby('Number of Cuisine').agg({'Aggregate rating': 'mean', 'Votes': 'sum'}).reset_index()
rating_by_cuisines['Aggregate rating'] = rating_by_cuisines['Aggregate rating'].round(2)
rating_by_cuisines
```

```python
# Plotting a bubble plot
plt.figure(figsize=(12, 6))
plt.scatter(x=rating_by_cuisines['Number of Cuisine'], y=rating_by_cuisines['Aggregate rating'], s=rating_by_cuisines['Votes']/250, alpha=0.5, c='skyblue', edgecolors='black')
plt.xlabel('Number of Cuisines Served')
plt.ylabel('Mean User Rating')
plt.title('Effect of Number of Cuisines on User Rating (Bubble Plot)')

# Adding annotations
x = np.array(rating_by_cuisines['Number of Cuisine'])
y = np.array(rating_by_cuisines['Aggregate rating'])

for index,value in zip(x,y):
  plt.annotate(value , (index,value) , ha='right')


plt.show()
```

**Explanation of using perticular plot for this question :**

Here I use Bubble Plot to visualize the average user rating, number of cuisines, and the number of votes in a single plot. Bubble plots are effective for displaying three dimensions of data. In this case, the x-axis represents the number of cuisines, the y-axis represents the mean user rating, and the size of the bubbles represents the number of votes. This allows for a more comprehensive understanding of the data. The bubble plot not only conveys the relationship between the number of cuisines and mean user rating but also provides insights into the popularity of each category based on the number of votes. Larger bubbles indicate more votes, giving a sense of the significance and general acceptance of restaurants with a certain number of cuisines.

By combining these two visualizations we get a layered understanding of the data, catering to both the average user rating and the popularity (measured by the number of votes) of restaurants serving different numbers of cuisines. This approach enhances the overall comprehension of the relationship between the variables being analyzed.

**Analysis**

The bar plot reveals a general trend where the average user rating tends to increase with the number of cuisines served. This suggests that offering a diverse range of cuisines might be positively correlated with user satisfaction.

The bubble plot incorporates the number of votes, providing a more nuanced view. Larger bubbles indicate higher vote counts, suggesting greater popularity.
Restaurants with a moderate number of cuisines (3 to 6) tend to have both higher average user ratings and significant vote counts. This indicates that a balance in the variety of cuisines may contribute to both user satisfaction and popularity.


Observing the bubble plot, certain data points may stand out as outliers. These could represent restaurants with exceptionally high or low user ratings despite the number of cuisines served. Exploring these outliers can provide insights into exceptional cases or areas that might need attention.

By considering the bubble sizes, you can see how the number of votes is distributed across different categories of the number of cuisines. This provides an understanding of not only the average satisfaction but also the overall engagement and feedback from users.

The analysis suggests that there might be an optimal range of cuisines (3 to 6) where restaurants tend to achieve a good balance between high average ratings and substantial popularity.

The data points with a low number of cuisines but high average ratings and votes could be interesting to investigate further. These restaurants might have a specific niche or unique offering that resonates well with users.

 ## 1.3.  Average Cost of Restaurant

```python
# using the groupby function in Pandas to group your DataFrame (zm) by the 'Aggregate rating' column
# and then calculating the mean of the 'Average Cost for two' column for each group.

avaragecost = zm.groupby('Aggregate rating')['Average Cost for two'].mean()
avaragecost
```

```python
# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(avaragecost.index, avaragecost.values, marker='o', linestyle='-', color='b')
plt.title('Average Cost for Two vs. Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Average Cost for Two')
plt.grid(True)
plt.show()
```

**Explanation of using perticular plot for this question :**

Here I choose line plot to visualize the relationship between the aggregate rating and the average cost for two. This plot is effective for showing trends and patterns in the data over a continuous variable. The x-axis represents the aggregate rating, and the y-axis represents the average cost for two.Line plots are suitable for visualizing trends or patterns over a continuous variable. Here, it helps see how average cost changes concerning aggregate ratings.


```python
# Create a bar plot
plt.figure(figsize=(10,7))

avaragecost.plot.bar()
plt.title('Distribution of Ratings based on Average Cost')
plt.xlabel('Aggregate Rating')
plt.ylabel('Average Cost for Two')

plt.show()
```

**Explanation of using perticular plot for this question :**

Here I choose bar plot since it is used to display the distribution of average cost for two across different aggregate ratings. This plot helps compare the average cost for two for each aggregate rating category.The x-axis represents the aggregate rating, and the y-axis represents the average cost for two.Bar plots are effective for comparing the distribution of a categorical variable across different categories. In this case, it helps to see how average cost varies for each aggregate rating.

```python
# Create a new column for average cost categories
bins = [0, 50, 100, 150, float('inf')]
labels = ['Low Cost', 'Medium Cost', 'High Cost', 'Very High Cost']
zm['Cost Category'] = pd.cut(zm['Average Cost for two'], bins=bins, labels=labels, right=False)
zm[['Average Cost for two','Cost Category']]
```

```python
# Create a Box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Cost Category', y='Aggregate rating', data=zm, order=labels)
plt.title('User Rating Distribution across Average Cost Categories')
plt.xlabel('Average Cost Categories')
plt.ylabel('User Rating')
plt.show()
```

**Explanation of using perticular plot for this question :**

Here I use box plot also sinceit is employed to analyze the distribution of user ratings across different average cost categories. This plot is useful for understanding the spread of data, including median, quartiles, and potential outliers. The x-axis represents the average cost categories, and the y-axis represents the user ratings.Box plots are excellent for comparing the distribution of a continuous variable across different categories. Here, it helps to understand the spread of user ratings for various average cost categories.

**Analysis**

The line plot indicates a general trend that as the aggregate rating increases, the average cost for two tends to rise.
This suggests that higher-rated restaurants, on average, may have a higher cost for two people.

The bar plot illustrates the distribution of average cost for two across different aggregate ratings.
It reveals that higher aggregate ratings tend to have a wider range of average costs, indicating greater variability in cost among highly rated restaurants.

The box plot highlights the distribution of user ratings within different average cost categories.
It shows that very high-cost restaurants have a wider spread of user ratings, potentially indicating more diverse customer opinions.

The grouped data on average cost for two in different rating categories provides a detailed breakdown.
Higher aggregate ratings (e.g., 4.1 and above) are associated with very high average costs for two, while lower ratings (e.g., 1.8 and 1.9) have lower average costs.

The combination of average cost for two and cost categories shows the distribution of restaurants based on their average costs.
Most restaurants fall into the 'Very High Cost' and 'High Cost' categories, with only a few in the 'Low Cost' and 'Medium Cost' categories.

Higher-rated restaurants tend to have higher average costs, but there is also variability within each rating category. The distribution of user ratings across different cost categories further emphasizes the diversity in customer preferences and experiences.
